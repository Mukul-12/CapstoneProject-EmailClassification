```python
# Import the needed  modules...
import pandas as pd   
import json
```

# Define Path to Data File

```python
# we get the path at which our file emails.csv is stored...
from os import getcwd
path = f"{getcwd()}/Data Store/emails.csv"
```

# Data Engineering

```python
# Read data from emails.csv

df=pd.read_csv(path)           
```

```python
# Display first five rows of emails.csv file...

df.head()
```

```python
# Index of series having column name

df.keys()
```

```python
# Here 0 indicates number of ham mails and 1 indicates number of spam mails...

df['Prediction'].value_counts()
```

```python
# It describes various thing of our data like count, mean, standard deviation etc..

df.describe()
```

```python
# It give the index number of our data from start to end and also give the size of step...

df.index
```

```python
# It count the number of rows in a particular column has null value...

df.isnull().sum()
```

```python
# Here we store all the spam mails in one place...

spam=df[df['Prediction']==1]
spam.head()
```

```python
# Here we store all the mails in one place...

ham=df[df['Prediction']==0]
ham.head()
```

```python
# Here we get the number of spam and ham mails...

ham.shape, spam.shape
```

```python
# Here we take a sample of ham mails in equal quantity of spam mails...

ham=ham.sample(spam.shape[0]) 
ham.shape
```

```python
# Know, Quantity of ham mails is equal to spam mails...

ham.shape, spam.shape
```

```python
# Here we make a data contain both ham and spam mails in equal amount...

data = ham.append(spam, ignore_index=True)
data
```

```python
 # Save dataframe to JSON format...
    
data.to_json("data1.json")       
```

```python
# Load the JSON file... 

with open('data1.json') as json_file:
    data1 = json.load(json_file)
```

```python
# Writing data into disk...

with open('data1.json', 'w') as outfile:
    json.dump(data1, outfile)
```
